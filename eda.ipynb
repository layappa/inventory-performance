function to insert a datafreame into sql

```python
def ingest_df(df,table,engine):
    df.to_sql(table,con=engine,if_exists='replace',index=False)

```

inserting dataset into sql

```python
def load_row_data():
    start=time.time()
    for file in os.listdir('data'):
        if '.csv' in file :
            df=pd.read_csv('data/'+file)
            ingest_df(df,file[:-4],engine)
            logging.info(f"ingesting {file} in db")
    end=time.time()
    logging.info(f"------------ingestion complete--------------")
    total_time=(end-start)/60
    logging.info(f"total_time taken : {total_time} minustes")
if __name__=='__main__':
    load_row_data()
```
